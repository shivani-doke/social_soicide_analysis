```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(12, 6))

# Count the occurrences of each response for depression and sleep issues
depression_counts = df['18. How often do you feel depressed or down?'].value_counts().sort_index()
sleep_issues_counts = df['20. On a scale of 1 to 5, how often do you face issues regarding sleep?'].value_counts().sort_index()

# Plotting the depression data
plt.subplot(1, 2, 1)
depression_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Feeling Depressed or Down')
plt.xlabel('Scale (1-5)')
plt.ylabel('Number of People')

# Plotting the sleep issues data
plt.subplot(1, 2, 2)
sleep_issues_counts.plot(kind='bar', color='salmon')
plt.title('Frequency of Facing Sleep Issues')
plt.xlabel('Scale (1-5)')
plt.ylabel('Number of People')

plt.tight_layout()
plt.show()
```
